Reemplazar los valores NaN por 0

1. ¿Cómo han variado los precios del gas natural a lo largo del tiempo?

```python
plt.figure(figsize=(15, 10))
sns.lineplot(x='Date', y='Natural_Gas_Vol.', data=df)
plt.title('Natural Gas Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

2. ¿Existe alguna correlación notable entre los precios del petróleo crudo y los precios del gas natural?

```python
correlation = df[['Crude_oil_Price', 'Natural_Gas_Price']].corr()
print(correlation)

sns.scatterplot(x='Crude_oil_Price', y='Natural_Gas_Price', data=df)
plt.title('Relación entre el Precio del Petróleo Crudo y el Gas Natural')
plt.xlabel('Precio del Petróleo Crudo')
plt.ylabel('Precio del Gas Natural')
plt.show()
```

3. ¿Qué activo tiene la distribución de precios más amplia?

```python
plt.figure(figsize=(15, 10))
sns.histplot(x='Crude_oil_Price', data=df, bins=30, kde=True)
sns.histplot(x='Natural_Gas_Price', data=df, bins=30, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Análisis Predictivos

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

6. Predicción del Precio del Gas Natural

```python
x = df[['Crude_oil_Price']]
y = df[['Natural_Gas_Price']]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

plt.figure(figsize=(15, 10))
sns.lineplot(x=X_test.index, y=y_test.squeeze(), label='Actual')
sns.lineplot(x=X_test.index, y=pd.Series(y_pred.squeeze(), index=X_test.index), label='Predicted')
plt.ylabel('Natural Gas Price')
plt.xlabel('Index')
plt.legend()
plt.show()
```
